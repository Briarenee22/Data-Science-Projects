```python
# Handling missing categorical columns

data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])
data['Married'] = data['Married'].fillna(data['Married'].mode()[0])
data['Dependents'] = data['Dependents'].fillna(data['Dependents'].mode()[0])
```

```python
#Handling missing Numerical columns
data ['LoanAmount']. fillna(data['LoanAmount'].median(), inplace=True)
data ['Loan_Amount_Term']. fillna(data['Loan_Amount_Term'].median(), inplace=True)
data ['Credit_History']. fillna(data['Credit_History'].median(), inplace=True)
```

# **Exploratory Data** **Anlaysis**

```python
#Lets explore our categorical data first to visualize insights that can relate to loan status
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data.columns
```

```python
#finding the distribution of loan status
sns.countplot(data=data, x='Loan_Status', color ='lightcoral')

#add x and y labels
plt.title('Loan status distribution')
plt.xlabel('loan_status')
plt.ylabel('count')

#Show Plot
plt.show()
```

```python
#count plot visualization for credit history
sns.countplot(x='Credit_History', data=data, color='blue')

#adding x and y labels
plt.title('Loan approval rate by credit history')
plt.xlabel('credit history(0= no history, 1= has history)')
plt.ylabel('count')

# Show plot
plt.show()
```

**Loan status by marriage**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Group the data by 'Married' and 'Loan_Status', then count the occurrences
grouped_data = data.groupby(['Married', 'Loan_Status']).size().unstack()

# Plot the stacked bar chart
ax = grouped_data.plot(kind='bar', stacked=True, color=['blue', 'orange'])

# Add Title and Labels
plt.title('Loan Approval by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')

# Add numbers to the bars
ax.bar_label(ax.containers[0], label_type='center')
ax.bar_label(ax.containers[1], label_type='center')

# Show the plot
plt.show()
```

```python
# Assuming your DataFrame is named df
plt.figure(figsize=(8,6))
sns.countplot(x='Education', hue='Loan_Status', data=data, palette= 'bright')

plt.title('Loan Status by Education Level')
plt.xlabel('Education')
plt.ylabel('Count')

#Show Plot
plt.show()
```

# **Numerical Data**

```python
# Summary statistics for the numerical columns
print(data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']].describe())
```

```python
#Adjust figure
plt.figure(figsize=(6, 6))

# Create a brighter palette
bright_palette = sns.color_palette("Set2", 2)

sns.boxplot(x='Loan_Status', y='ApplicantIncome', data=data, palette=bright_palette, hue='Loan_Status', legend=False)
plt.title('Applicant Income vs Loan Status')

# Show plot
plt.show()
```

```python
#Create Histoplot
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='CoapplicantIncome', hue='Loan_Status', multiple='stack', bins=30, palette='bright', kde=True)
plt.title('Co-Applicant Income Distribution by Loan Status')
plt.xlabel('Co-Applicant Income')
plt.ylabel('Count')

#Show plot
plt.show()
```

```python
# Adjust the figure size
plt.figure(figsize=(8, 6))

# Create violinplot
sns.violinplot(x='Loan_Status', y='LoanAmount', data=data, palette='Set3', hue='Loan_Status', legend=False)
plt.title('Loan Amount Distribution by Loan Status (Violin Plot)')
plt.xlabel('Loan Status')
plt.ylabel('Loan Amount')

# Show plot
plt.show()
```
